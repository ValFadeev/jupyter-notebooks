## The Simplest Quantum System

Consider a quantum system that can exist in one of two states customarily labelled $0$ and $1$, respectively. Such a system is characterised by a two-component state-vector

$$\chi \equiv \begin{bmatrix}\psi(0)\\ \psi(1)\end{bmatrix} = \begin{bmatrix}\alpha \\ \beta \end{bmatrix},$$

where $\alpha$ and $\beta$ are complex numbers. Alternatively, $\chi$ can be written as a superposition of the system being in the states

$$|0\rangle = \begin{bmatrix}1 \\ 0\end{bmatrix}, \quad |1\rangle = \begin{bmatrix}0 \\ 1\end{bmatrix},$$

as follows

$$\chi = \alpha|0\rangle + \beta|1\rangle. \tag{1}\label{qubit}$$

Consequently, the probabilities of the system being found in either $|0\rangle$ or $|1\rangle$ are $|\alpha|^2$ and $|\beta|^2$, respectively, and we must have

$$|\alpha|^2 + |\beta|^2.$$

The form of the state vector given by \eqref{qubit} is often referred to as the _qubit_. Evolution of our system will be described by a Hamiltonian represented by a $2\times 2$ matrix

$$\mathcal{H}=\begin{bmatrix}H_{00} & H_{01} \\ H_{10} & H_{11}\end{bmatrix}.$$

The hermiticity requirement dictates that we must set $H^{*}_{01} = H_{10} = a_1 + ia_2$ where $a_1$, $a_2$ are real. Furthermore, without loss of generality we can set $H_{00} = a_0 + a_3$ and $H_{00} = a_0 - a_3$ (again, $a_0$, $a_3$ being real numbers).

$$\mathcal{H}=\begin{bmatrix}a_0 + a_3 & a_1 - ia_2 \\ a_1+ia_2 & a_0-a_3\end{bmatrix}=a_0\sigma_0+a_1\sigma_x+a_2\sigma_y+a_3\sigma_z = a_0\sigma_0 + \vec{a}\cdot\vec{\sigma}.$$

Here $\sigma_x, \sigma_y, \sigma_z$ are the standard Pauli matrices

$$
\sigma_x=\begin{bmatrix}0 & 1 \\ 1 & 0\end{bmatrix}, \quad \sigma_y=\begin{bmatrix}0 & -i \\ i & 0\end{bmatrix}, \quad \sigma_z=\begin{bmatrix}1 & 0 \\ 0 & -1\end{bmatrix},
$$

and $\sigma_0$ is the $2\times 2$ identity matrix. Using the identity

$$
(\vec{\sigma}\cdot\vec{a})(\vec{\sigma}\cdot\vec{b}) = \vec{a}\cdot\vec{b} + i\vec{\sigma}\cdot\vec{a}\times\vec{b} \label{eq1}\tag{2}
$$

Setting $\vec{a}=\vec{b}=\vec{n}$ where $\vec{n}$ is a unit vector we find

$$
(\vec{\sigma}\cdot\vec{n})^2 = 1.
$$

The interpretation of this is that the eigenvalues of $\vec{\sigma}\cdot{n}$ are always $\pm 1$. Using the same identity \eqref{eq1} we can find the eigenenergies of the system

$$
\begin{align}
\mathcal{H}|\chi\rangle &= E|\chi\rangle, \\
\vec{\sigma}\cdot\vec{a}|\chi\rangle &= (E-a_0)|\chi\rangle \label{eq2}\tag{3}.
\end{align}
$$

Now multiply both sides by $\vec{\sigma}\cdot\vec{a}$ and observe that $(\vec{\sigma}\cdot\vec{a})^2=\vec{a}\cdot\vec{a}+i\vec{\sigma}\cdot\underbrace{\vec{a}\times \vec{a}}_{=\vec{0}} = |\vec{a}|^2$ to obtain

$$
|\vec{a}|^2|\chi\rangle=(E-a_0)\vec{\sigma}\cdot\vec{a}|\chi\rangle. \label{eq3}\tag{4}
$$

Comparing \eqref{eq2} and \eqref{eq3} we find

$$
E = a_0 \pm |\vec{a}|
$$

## Unitary Geometry

Writing the components of a unit vector $\vec{n}$ in spherical coordinates

$$
\vec{n}=(n_x, n_y, n_z) = (\sin{\theta}\cos{\phi}, \sin{\theta}\sin{\phi}, \cos{\theta}),
$$

we find

$$
\vec{\sigma}\cdot\vec{n} = \begin{bmatrix}\cos{\theta} & \sin{\theta}e^{-i\phi} \\ \sin{\theta}e^{i\phi} & -\cos{\theta}\end{bmatrix} \tag{5}\label{spinor}
$$

Eigenvectors of \eqref{spinor} corresponding to the eigenvalues +1 and -1, respectively are called _spinors_ and can be written as follows

$$
\begin{align}
|\chi^{+}\rangle &= \cos\frac{\theta}{2}|0\rangle + \sin\frac{\theta}{2}e^{i\phi}|0\rangle \\
|\chi^{-}\rangle &= \sin\frac{\theta}{2}|0\rangle - \cos\frac{\theta}{2}e^{i\phi}|0\rangle
\end{align}
$$

Adding a global phase factor $e^{i\gamma}$ the most general form of a spinor is found

$$
|\chi^{+}\rangle = e^{i\gamma}\left(\cos\frac{\theta}{2}|0\rangle + \sin\frac{\theta}{2}e^{i\phi}|0\rangle\right) \tag{6}\label{spinor_gen}
$$

The angles $\theta$ and $\phi$ determine the representation of a spinor on the Bloch sphere. The qubit ([1](#mjx-eqn-qubit)) can be written in the form \eqref{spinor_gen} as follows

$$
|\chi\rangle = \begin{bmatrix}|\alpha|e^{i\phi_{\alpha}} \\ |\beta|e^{i\phi_{\beta}} \end{bmatrix} = e^{i\phi_{\alpha}}\begin{bmatrix}|\alpha| \\ |\beta|e^{i(\phi_{\beta}-\phi_{\alpha})} \end{bmatrix}. 
$$

Hence, it suffices to set

$$
\begin{align}
\gamma &= \phi_{\alpha} \\
\theta &= 2\arctan\left(\frac{\sqrt{1-|\alpha|^2}}{|\alpha|}\right)\\
\phi &= \phi_{\alpha} - \phi_{\beta}
\end{align}
$$

In particular states $|0\rangle$ and $|1\rangle$ are identified with the North and South poles of the Bloch sphere

$$
\begin{align}
|0\rangle &= |\chi^{+}(\theta=0, \phi, \gamma) \\
|1\rangle &= |\chi^{+}(\theta=\pi, \phi, \gamma)
\end{align}
$$


# Quantum logic gates as rotations

### The Hadamard matrix

The Hadamard matrix $H$ is given by

$$
H=\frac{1}{\sqrt{2}}\begin{bmatrix}1 & 1 \\ 1 & -1 \end{bmatrix}
$$

One of its eigenvectors can be shown to be 

$$
|\chi\rangle = \cos\frac{\pi}{8}|0\rangle + \sin\frac{\pi}{8}|1\rangle,
$$

whereby we find that $\theta=\frac{\pi}{4}$ and $\phi=0$. Hence this eigenvector lies in the $xz$ plane and is rotated halfway between the north pole and the point where the positive $x$-axis intersects the Bloch sphere.

```python
coords = [pi/4,0,1] # [Theta, Phi, Radius]
plot_bloch_vector_spherical(coords) # Bloch Vector with spherical coordinates
```

### The Spin Flip Matrix

It is easy to check that the matrix

$$
M = \begin{bmatrix} 0 & -e^{-i\phi} \\ e^{i\phi} & 0 \end{bmatrix}
$$

changes $|\chi|^{-}\rangle$ to $|\chi^{+}\rangle$. $M$ can also be written as a product

$$
M = \begin{bmatrix} 1 & 0 \\ 1 & e^{i\phi} \end{bmatrix} \begin{bmatrix} 0 & -1 \\ 1 & 0 \end{bmatrix} \begin{bmatrix} 1 & 0 \\ 1 & e^{-i\phi} \end{bmatrix} = e^{i\frac{\pi}{2}}P(\phi)\sigma_yP(-\phi),
$$

where

$$
P(\phi) = \begin{bmatrix} 1 & 0 \\ 0 & e^{i\phi} \end{bmatrix}
$$

is the phase-shift matrix. Its special cases frequently used in the design of quantum logic gates are the $S$ matrix

$$
S = \begin{bmatrix} 1 & 0 \\ 0 & i \end{bmatrix}
$$

and the $T$ matrix

$$
T = \begin{bmatrix} 1 & 0 \\ 0 & e^{i\frac{\pi}{4}} \end{bmatrix}
$$
